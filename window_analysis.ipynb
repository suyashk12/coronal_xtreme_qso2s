```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

III Zw 77

```python
obj_fits_zw = fits.open("spec-1171-52753-0076.fits")
```

```python
spec_obj_zw = obj_fits_zw[2].data
spec_hdr_zw = obj_fits_zw[2].header

n_fields_zw = spec_hdr_zw['tfields']

redshift_idx_zw = 0

for i in range(1, n_fields_zw):
    if(spec_hdr_zw['ttype' + str(i)] == 'Z'):
        redshift_idx_zw = i
        break

redshift_zw = spec_obj_zw[0][redshift_idx_zw-1]
```

```python
redshift_zw
```

```python
#Obtain spectroscopic information
spec_zw = list(obj_fits_zw[1].data)
spec_zw = [list(spec_zw[i]) for i in range(len(spec_zw))]
spec_zw = np.array(spec_zw)

wav_zw = 10**spec_zw[:,1]
flux_zw = spec_zw[:,0]
```

```python
wav_shifted_zw = wav_zw/(1+redshift_zw)
```

```python
fig, ax = plt.subplots(1, figsize=(15, 7))

ax.plot(wav_shifted_zw, flux_zw)
ax.set_xlabel('Wavelength (Angstroms)')
ax.set_ylabel('Flux Density (1e-17 ergs/s/cm^2/A)')
ax.set_title('III Zw 77 Spectrum')
```

Construct dictionary

```python
qso_names = ['III Zw 77', 'Mrk 1388', 'Rose CLiF', 'New CLiF']

info_dict = {'III Zw 77': [wav_shifted_zw, flux_zw],
             'Mrk 1388': [wav_shifted_mrk, flux_mrk],
             'Rose CLiF': [wav_shifted_rose, flux_rose],
             'New CLiF': [wav_shifted_new, flux_new]}

line_dict = {}

eps = 10
```

[Ne_V] 3425

```python
ne_v_3425_cent = 3425

ne_v_3425_left_width = 15
ne_v_3425_left_space = 13
ne_v_3425_left_cont = 30

ne_v_3425_right_width = 17
ne_v_3425_right_space = 20
ne_v_3425_right_cont = 30

ne_v_3425_min_wav = ne_v_3425_cent - ne_v_3425_left_width - ne_v_3425_left_space - ne_v_3425_left_cont
ne_v_3425_max_wav = ne_v_3425_cent + ne_v_3425_right_width + ne_v_3425_right_space + ne_v_3425_right_cont
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(9*2, 6*2))

for i in range(2):
    for j in range(2):
        obj_name = qso_names[2*i + j]
        wav_shifted = info_dict[obj_name][0]
        flux = info_dict[obj_name][1]
        axes[i,j].plot(wav_shifted, flux)
        axes[i,j].set_title(obj_name + ' Spectrum')
        axes[i,j].axvspan(ne_v_3425_min_wav,
                          ne_v_3425_min_wav+ne_v_3425_left_cont,
                          alpha=.4,color='violet')
        axes[i,j].axvspan(ne_v_3425_cent-ne_v_3425_left_width,
                          ne_v_3425_cent+ne_v_3425_right_width,
                          alpha=.2,color='green')
        axes[i,j].axvspan(ne_v_3425_max_wav-ne_v_3425_right_cont,
                          ne_v_3425_max_wav,alpha=.3,color='red')
        axes[i,j].set_xlabel('Wavelength (Angstroms)')
        axes[i,j].set_ylabel('Flux Density (1e-17 ergs/s/cm^2/A)')
        axes[i,j].set_xlim(ne_v_3425_min_wav-eps, ne_v_3425_max_wav+eps)

fig.suptitle('[NeV] 3425 Windows')
```

```python
line_dict['[Ne_V] 3425'] = [ne_v_3425_cent, ne_v_3425_left_width, ne_v_3425_left_space, ne_v_3425_left_cont,
                         ne_v_3425_right_width, ne_v_3425_right_space, ne_v_3425_right_cont]

line_dict
```
